# Word Embeddings with Gensim
This notebook demonstrates how to load pre-trained **Word2Vec** and **GloVe** models using `gensim` and generate embeddings for new text data.

### 1. Imports and Setup

```python
import numpy as np
import gensim.downloader as api
import gensim
from gensim.models import KeyedVectors

print("Gensim version:", gensim.__version__)
```

### 2. Loading Pre-trained Models
Note: These files are large and will take a moment to download on the first run.

```python
print("Loading Word2Vec (Google News 300)...")
w2v_model = api.load("word2vec-google-news-300")

print("Loading GloVe (Wiki GigaWord 300)...")
glove_model = api.load("glove-wiki-gigaword-300")

print("Models loaded successfully!")
```

### 3. Function to Generate Sentence Embeddings
We generate phrase-level vectors by calculating the mean of the word vectors present in the model's vocabulary.

```python
def get_phrase_embedding(phrase, model):
    """
    Generates a single vector for a phrase by averaging word vectors.
    """
    words = phrase.lower().split()
    
    # Extract vectors for words that exist in the model vocabulary
    valid_vectors = [model[word] for word in words if word in model]
    
    if not valid_vectors:
        return np.zeros(model.vector_size)
    
    return np.mean(valid_vectors, axis=0)

# Test on a single sentence
test_sentence = "The quick brown fox jumps over the lazy dog"
sample_vector = get_phrase_embedding(test_sentence, w2v_model)
print(f"Vector shape for test sentence: {sample_vector.shape}")
```

### 4. Processing New Data
Applying the embedding logic to a collection of new strings.

```python
new_texts = [
    "Machine learning is fascinating.",
    "Artificial intelligence will change the world.",
    "I love processing natural language data.",
    "Python is the best language for data science."
]

print("Generating Word2Vec embeddings...")
w2v_embeddings = np.array([get_phrase_embedding(text, w2v_model) for text in new_texts])

print("Generating GloVe embeddings...")
glove_embeddings = np.array([get_phrase_embedding(text, glove_model) for text in new_texts])

print(f"Word2Vec Matrix Shape: {w2v_embeddings.shape}")
print(f"GloVe Matrix Shape: {glove_embeddings.shape}")
```
